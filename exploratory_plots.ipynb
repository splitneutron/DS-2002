```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
modern_df = pd.read_csv("https://raw.githubusercontent.com/Guillaumeserres31/Twitter-Sentiment-Analysis-on-Generative-AI-Adoption/refs/heads/main/Dataset/GenerativeAI%20tweets.csv")

print("Modern AI Dataset Statistics:")
print(modern_df.describe())
# print("\nPre AI Dataset Statistics:")
# print(pre_df2.describe())
```

```python
modern_df.head()
```

```python
modern_df['Datetime'] = pd.to_datetime(modern_df['Datetime'])
tweets_by_day = modern_df.groupby(modern_df['Datetime'].dt.date).size().reset_index(name='tweet_count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=tweets_by_day, x='Datetime', y='tweet_count')
plt.title('Number of Tweets Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
modern_df['word_count'] = modern_df['Text'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(10, 6))
sns.histplot(data=modern_df, x='word_count', bins=50, kde=True)
plt.title('Distribution of Word Count in Tweets')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()
```
